```python
from geopy.geocoders import Nominatim

# Initialize the geocoder
geolocator = Nominatim(user_agent="my_geocoder")

# Define the location you want to geocode
location_name = "Almafuerte , Argentina"

# Geocode the location
location = geolocator.geocode(location_name)

if location:
    print("Location Details:")
    print(f"Name: {location}")
    print(f"Latitude: {location.latitude}")
    print(f"Longitude: {location.longitude}")
else:
    print("Location not found or geocoding failed.")
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

df = pd.read_csv('/home/glue_user/workspace/jupyter_workspace/ITBA/datascience_final_project/llcc_without_lat_and_long.tsv', delimiter='\t')


def get_lat_long(location_name, partido, provincia):
    location_name = location_name + " , " + partido + " , " + provincia + " , Argentina"
    geolocator = Nominatim(user_agent="my_geocoder")
    try:
        location = geolocator.geocode(location_name)
    except:
        return None, None, None
    if location:
        print(location_name, location.latitude, location.longitude, location)
        return location.latitude, location.longitude, location
    else:
        print(None, None, None)
        return None, None, None
df['Latitude'], df['Longitude'], df['location'] = zip(*df.apply(lambda row: get_lat_long(row['Localidad'], row['Partido'], row['Provincia']), axis=1))

df.to_csv('/home/glue_user/workspace/jupyter_workspace/ITBA/datascience_final_project/llcc_with_lat_and_log.tsv', sep='\t', index=False)
```
